```python
import pandapower as pp
net=pp.create_empty_network(sn_mva=100)
```

```python
bus0=pp.create_bus(net, name="Busbar 0", vn_kv=0, type="n")
bus1=pp.create_bus(net, name="Busbar 1", vn_kv=22, type="b")
bus2=pp.create_bus(net, name="Busbar 2",vn_kv=11, type="n")
bus3 = pp.create_bus(net, name="Busbar 3", vn_kv=220, type="b")
bus4 = pp.create_bus(net, name="Busbar 4", vn_kv=220, type="b")
bus5 = pp.create_bus(net, name="Busbar 5", vn_kv=110, type="b")
bus6 = pp.create_bus(net, name="Busbar 6", vn_kv=110, type="b")
```

```python
net.bus
```

```python
trafo1=pp.create_transformer_from_parameters(net, bus3, bus1, name="T1", sn_mva=50, vn_hv_kv=220, vn_lv_kv=22, vkr_percent=45, vk_percent=0, pfe_kw=0, i0_percent=0)
trafo2=pp.create_transformer_from_parameters(net, bus4, bus2, name="T2", sn_mva=40, vn_hv_kv=220, vn_lv_kv=11, vkr_percent=60, vk_percent=0, pfe_kw=0, i0_percent=0)
trafo3=pp.create_transformer_from_parameters(net, bus5, bus1, name="T3", sn_mva=40, vn_hv_kv=110, vn_lv_kv=22, vkr_percent=53, vk_percent=0, pfe_kw=0, i0_percent=0)
trafo4=pp.create_transformer_from_parameters(net, bus6, bus2, name="T4", sn_mva=40, vn_hv_kv=110, vn_lv_kv=11, vkr_percent=44, vk_percent=0, pfe_kw=0, i0_percent=0)
```

```python
net.trafo
```

```python
line1=pp.create_line_from_parameters(net, bus3, bus4, length_km=1, r_ohm_per_km=0, x_ohm_per_km=48.4, c_nf_per_km=0, max_i_ka=0.23, name="Line1")
line2=pp.create_line_from_parameters(net, bus5, bus6, length_km=1, r_ohm_per_km=0, x_ohm_per_km=65.43, c_nf_per_km=0, max_i_ka=0.23, name="Line2")
```

```python
net.line
```

```python
pp.create_gen(net, bus1, p_mw=54, sn_mva=90, vn_kv=22, name="generator")
net.gen
```

```python
pp.create_load_from_cosphi(net, bus2, sn_mva=57, cos_phi=0.6, scaling=1.108, const_z_percent=0, mode="overexcited", name="load")
net.load
```

```python
impedance=pp.create_impedance(net, bus0, bus1, r_pu=0.03, x_pu=0.18, sn_mva=90, rft_pu=None, xft_pu=None)
net.impedance
```

```python
pp.create_ext_grid(net, bus1, name="ext. grid", vm_pu=1, va_degree=50)
net.ext_grid
```

```python
pp.runpp(net)
net.res_bus
```
