```python
import pandapower as pp
net=pp.create_empty_network()
```

```python
bus1=pp.create_bus(net, name="Busbar 1", vn_kv=22.011, type="n")
bus2=pp.create_bus(net, name="Busbar 2",vn_kv=2.992, type="n")
bus3 = pp.create_bus(net, name="Busbar 3", vn_kv=207.68, type="b")
bus4 = pp.create_bus(net, name="Busbar 4", vn_kv=209.88, type="b")
bus5 = pp.create_bus(net, name="Busbar 5", vn_kv=111.1, type="b")
bus6 = pp.create_bus(net, name="Busbar 6", vn_kv=99, type="b")
```

```python
net.bus
```

```python
trafo1=pp.create_transformer_from_parameters(net, bus3, bus1, name="T1", sn_mva=50, vn_hv_kv=220, vn_lv_kv=22)
trafo2=pp.create_transformer_from_parameters(net, bus4, bus2, name="T2", sn_mva=40, vn_hv_kv=220, vn_lv_kv=11)
trafo3=pp.create_transformer_from_parameters(net, bus5, bus1, name="T3", sn_mva=40, vn_hv_kv=110, vn_lv_kv=22)
trafo4=pp.create_transformer_from_parameters(net, bus6, bus2, name="T4", sn_mva=40, vn_hv_kv=110, vn_lv_kv=11)
```

```python
net.trafo
```

```python
line1=pp.create_line(net, bus3, bus4, length_km=1, std_type="N2XS(FL)2Y 1x240 RM/35 64/110 kV",  name="Line1")
line2=pp.create_line(net, bus5, bus6, length_km=1, std_type="NA2XS2Y 1x240 RM/25 12/20 kV", name="Line2")
```

```python
net.line
```

```python
pp.create_gen(net, bus1, p_mw=54, sn_mva=90, vn_kv=22, name="generator")
net.gen
```

```python
pp.create_load_from_cosphi(net, bus2, sn_mva=57, cos_phi=0.6, mode="underexcited", name="load")
net.load
```

```python
pp.create_ext_grid(net, bus1, vm_pu=1, va_degree=50)
net.ext_grid
```

```python
pp.runpp(net)
net.res_bus
```
